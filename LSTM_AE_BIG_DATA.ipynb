```python
from keras.layers import *
from keras.layers import LSTM, RepeatVector
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from keras import Sequential
import keras as kr


boat_csv = pd.read_csv("Data/Boat_nominal_data/Boat_nominal_long.csv")
boat_csv = boat_csv.drop(columns=["Unnamed: 0"])
scaler = StandardScaler()
normal_data = scaler.fit_transform(boat_csv)
print(normal_data.shape)
```

```python
model = Sequential()
encoder = Sequential(name="encoder")
encoder.add(Dense(units=663,activation='relu'))
encoder.add(Dense(300, activation="relu"))
encoder.add(Dense(30, activation='relu'))

decoder = Sequential()
decoder.add(Dense(30, activation='relu'))
decoder.add(Dense(300, activation="relu"))
decoder.add(Dense(boat_csv.shape[1], activation='relu'))

model.add(encoder)
model.add(decoder)

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(normal_data,normal_data,verbose=1,epochs=10, validation_split=0.1)
### END AUTOENCODER

model.save("Models/Boat_nominal.model")

print("End module")
```

```python
nom_model = kr.models.load_model("Models/Boat_nominal.model")
print("Nominal MODEL LOADED")
```
